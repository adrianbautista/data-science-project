```python
import os
import foursquare as fs
```

```python
client = fs.Foursquare(client_id=os.environ['FOURSQUARE_DS_CLIENT'], client_secret=os.environ['FOURSQUARE_DS_CLIENT_SEC'])
```

```python
result = client.venues(rest_df['foursquare_id'].iloc[0])
```

```python
print result['venue']['location']
```

```python
def get_lat_long(id):
    try:
        result = client.venues(id)
        if 'venue' in result and 'location' in result['venue'] and ('lat' in result['venue']['location'] and 'lng' in result['venue']['location']):
            print 'geolocated: ', id
            return (result['venue']['location']['lat'], result['venue']['location']['lng'])
    except:
        print 'could not geolocate ', id
        
rest_df['lat_long'] = rest_df.foursquare_id.apply(get_lat_long)
```
